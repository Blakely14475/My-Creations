```python
#alright let's try an online tutorial for MultiIndexing
import pandas as pd
stocks = pd.read_csv('http://bit.ly/smallstocks') #okay that's pretty cool. import a csv from
#a url
```

```python
stocks
```

```python
stocks.index
```

```python
stocks.groupby('Symbol').Close.mean()
```

```python
ser = stocks.groupby(['Symbol', 'Date']).Close.mean() #for each instance in which the symbol & date
#show that average Close. this would be more useful if each Symbol had more than one Close value
#for a given date
```

```python
#note that this is a series with a MultiIndex
#note that this is a series with a MultiIndex
print(type(ser))
ser.index #i.e. our series has multiple dimensions
```

```python
#if we command Pandas to unstack this MultiIndex series, it will become a dataframe
df = ser.unstack()
print(type(df))
df
```

```python
#there's a simpler way to do all of this though
stockdf = stocks.pivot_table(values='Close', index='Symbol', columns='Date')
stockdf
```

```python
ser.loc['AAPL'] #in a multi-index series, we can select all information related to the element
#we pass it
```

```python
#we can get even MORE specific with the data we want
ser.loc['AAPL', '2016-10-03']
```

```python
#let's say we want all info from a date
ser.loc[:, '2016-10-03'] #Note that in python, : here serves to mean "All"
```

```python
#In MultiIndexing, all of these commands work just the same on the DataFrame as they work on
#the unstacked series

stockdf.loc['AAPL'] #BEHOLD!
```
